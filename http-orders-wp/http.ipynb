```python
'''
检查pcap文件中是否包含jquery的http请求
'''
def check_whether_contains_jquery(filenames):
    tshark_cmd = "tshark -r {} -Y 'http.request.uri contains \"jquery\"' | wc -l".format(filenames)
    r = os.popen(tshark_cmd)
    result = r.read()
    r.close()
    print(tshark_cmd)
    return int(result)
check_whether_contains_jquery('../data/http_flows/http_stage2-2425.pcap')
```

```python
'''
将指定文件复制到指定文件夹
'''
def copyFile(src, dst):
    if dst.endswith('/'):
        cpCmd = "cp {} {}".format(src, dst)
    else:
        cpCmd = "cp {} {}/".format(src, dst)
    print(cpCmd)
    os.system(cpCmd)
```

```python
total = 0
for filename in filenames:
    result = check_whether_contains_jquery(filename)
    if result==0:
        continue
    dstPath = "../data/http_jquery/"
    copyFile(filename, dstPath)
    total += 1
```

```python
def parseJqueryPcapByTshark(filename):
    cmd = 'tshark -r {} -Y "http" -T json -e http.date -e http.request.method -e http.file_data -e http.request.uri >> ./jquery_parse_json/{}.json'.format(filename, filename.split('/')[-1])
    print(cmd)
    os.system(cmd)    
```

```python
for filename in jqueryFiles:
    parseJqueryPcapByTshark(filename)
```

```python
def parseJqueryJson(filename):
    with open(filename, 'r') as f:
        data = json.loads(f.read())
    result = []
    for i, item in enumerate(data):
#         ipdb.set_trace()
        item = item['_source']['layers']
        if 'http.file_data' in item.keys() and 'http.request.method' not in item.keys():
            jsdata = item['http.file_data']
            payload = re.findall(r'.*?,P="(.*?)".*?', jsdata[0])
            if len(payload) > 0:
                item['http.file_data'] = payload[0]
            else:
                item['http.file_data'] = ""
            result.append(item)
        elif 'http.file_data' in item.keys() and 'http.request.method' in item.keys():
            item['http.file_data'] = item['http.file_data'][0]
            result.append(item)
    return result
```

```python
allResult = []
jqueryJsonPath = "./jquery_parse_json/"
jqeuyFiles = [os.path.join(jqueryJsonPath, x) for x in os.listdir(jqueryJsonPath)]
for file in jqeuyFiles:
    print("parse {}".format(file))
    allResult += parseJqueryJson(file)
```
